```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from random import choice, randint


movies=pd.read_csv('movie.csv')
print(movies.columns)
print(movies)
```

**Numpy**

```python
ratings=np.array(movies['vote_average'])
title=np.array(movies['title'])

avg_rating=np.mean(ratings)
min_rating=np.min(ratings)
max_rating=np.max(ratings)

print(avg_rating)
print(min_rating)
print(max_rating)

```

**Pandas task**

```python
import pandas as pd

max_rated_movies=movies[movies['vote_average']==max_rating]['title'].values[0]
min_rated_movies=movies[movies['vote_average']==min_rating]['title'].values[0]

print(f"Maximum Rating: {max_rating} for the movie: '{max_rated_movies}'")
print(f"Minimum Rating: {min_rating} for the movie: '{min_rated_movies}'")
```

**Matplotlib graph**



```python
import pandas as pd
import matplotlib.pyplot as plt

movies = pd.read_csv('movie.csv')
movies['release_year'] = pd.to_datetime(movies['release_date']).dt.year

movies_per_year = movies.groupby('release_year')['title'].count().reset_index()

movies_per_year.columns = ['Release Year', 'Number of Movies']

plt.figure(figsize=(20, 10))
plt.bar(movies_per_year['Release Year'], movies_per_year['Number of Movies'], color='pink')

plt.title('Number of Movies Released per Year', fontsize=25, color='Black',)
plt.xlabel('Release Year', fontsize=20, color='Black')
plt.ylabel('Number of Movies', fontsize=20, color='Black')

plt.grid(True)
plt.show()

```

**Stem PLot**

```python
import pandas as pd
import matplotlib.pyplot as plt

movies = pd.read_csv('movie.csv')
movies['release_year'] = pd.to_datetime(movies['release_date']).dt.year

movies_per_year = movies.groupby('release_year')['title'].count().reset_index()
movies_per_year.columns = ['Release Year', 'Number of Movies']

plt.figure(figsize=(20, 10))
plt.stem(movies_per_year['Release Year'], movies_per_year['Number of Movies'], linefmt='b-', markerfmt='bo', basefmt='r-')
plt.title('Number of Movies Released per Year (Stem Plot)', fontsize=25, color='Black')
plt.xlabel('Release Year', fontsize=20, color='Black')
plt.ylabel('Number of Movies', fontsize=20, color='Black')
plt.grid(True)
plt.show()
```

**Violin graph**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

movies = pd.read_csv('movie.csv')
movies['release_year'] = pd.to_datetime(movies['release_date']).dt.year

movies_per_year = movies.groupby('release_year')['title'].count().reset_index()
movies_per_year.columns = ['Release Year', 'Number of Movies']

plt.figure(figsize=(20, 10))
plt.violinplot([movies_per_year['Number of Movies']], showmeans=True)
plt.title('Distribution of Number of Movies Released per Year (Violin Plot)', fontsize=25, color='Black')
plt.xlabel('Release Year', fontsize=20, color='Black')
plt.ylabel('Number of Movies', fontsize=20, color='Black')
plt.xticks(ticks=[1], labels=['Number of Movies'])
plt.grid(True)
plt.show()
```
